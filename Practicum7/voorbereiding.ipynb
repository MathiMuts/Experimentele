# grafiek voor voorbereiding
Hieronder wordt een grafiek gegenereerd voor de voorbereiding van Practicum 7. Het verloop van de reactie wordt geplot voor een 0de, 1ste en 2de orde reactie.

```python
import numpy as np
import fit_classes as fp
import matplotlib.pyplot as plt
```

```python
k=1
R0 = 1

t = np.linspace(0, 10, 1000)
m0 = -k*t + R0
m1 = R0*np.exp(-k*t)
m2 = 1/(k*t + R0)

plt.figure(figsize=(10, 6))

plt.plot(t, m0, label='nulde orde ($m=0$)')
plt.plot(t, m1, label='eerste orde ($m=1$)')
plt.plot(t, m2, label='tweede orde ($m=2$)')

plt.xlim(0, 8)
plt.ylim(0, 1.5)

plt.title('Verloop van $[R](t)$ voor $m = 0, 1, 2$', fontsize=16)
plt.xlabel("$k't$", fontsize=14)
plt.ylabel('$[R]_0$', fontsize=14)
plt.legend(fontsize=12)
plt.show()
```
